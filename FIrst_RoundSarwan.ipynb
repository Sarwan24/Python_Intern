# First Problem

```python
n = int(input())

interval = []

for i in range(n):
    start, end = map(int, input().split())
    interval.append([start, end])

interval.sort()

merged = []

for i in interval:
    if not merged or merged[-1][1] < i[0]:
        merged.append(i)
    else:
        merged[-1][1] = max(merged[-1][1], i[1])

for start, end in merged:
    print(start, end)
```

# Second Problem

```python
n = int(input())

arr = list(map(int, input().split()))

k = int(input())

sums = sum(arr[:k])
max_sum = sums

for i in range(k, n):
    sums = sums + arr[i] - arr[i-k]
    if sums > max_sum:
        max_sum = sums

print(max_sum)
```

# Problem 3

```python
st = input().strip()

mx = 0

for i in range(len(st)):
    for j in range(i, len(st)):
        substring = st[i:j+1]

        if len(substring) == len(set(substring)):
            if len(substring) > mx:
                mx = len(substring)

print(mx)
```
